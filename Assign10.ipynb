```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
kmedoids = KMedoids(n_clusters=5, metric='manhattan', random_state=42)
kmedoids.fit(X)
```

```python
labels = kmedoids.labels_
medoids = kmedoids.cluster_centers_
colors = ['red', 'blue', 'green', 'orange', 'purple']
```

```python
for i in range(5):
    plt.scatter(X[labels == i, 0], X[labels == i, 1],
                s=50, c=colors[i], label=f'Cluster {i+1}', alpha=0.6)
plt.scatter(medoids[:, 0], medoids[:, 1],
            c='black', marker='X', s=200, label='Medoids')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Customer Segmentation using K-Medoids (Colored Clusters)")
plt.legend()
plt.show()
```
